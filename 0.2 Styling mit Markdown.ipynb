## Markdown im Jupyter Notebook

Neben Kommentaren kannst du wichtige Infos auch in Markdown-Zellen, hier im Jupyter Notebook, hinterlegen. Alle Texte, die du hier bisher gelesen hast, Erklärungen und Aufgabenstellungen, sind jeweils in Markdown-Zellen hinterlegt. Code hingegen wird in Code-Zellen ausgeführt. Markdown-Zellen sind Zellen, die von Python nicht als Code gelesen werden und in die du somit ganz normalen Fließtext schreiben kannst.

### 1. Es gibt drei Arten, eine neue Markdown-Zelle anzulegen bzw. eine Code- in eine Markdown-Zelle umzuwandeln:  
<br>
(Eine neue Zelle öffnest du zunächst mit dem **+**-Zeichen ganz oben.)
1. Du klickst im herunterklappenden Menü/Dropdown-Menü  (das ist die grau hinterlegte Pfeilspitze nach unten, auf der gleichen Zeile wie <b>+</b> neben <b>Code</b>) und wählst <b>Markdown</b> aus.
2. Du drückst in der neuen Zelle auf der Tastatur <font color = green>Esc</font> plus <font color = green>M</font>.
3. Aus dem <b>Cell</b>-Auswahlmenü ganz oben wählst du <b>Cell Type</b> aus und dann <b>Markdown</b>.

 <div class="alert alert-block alert-info">
<font size="3"><b>Tipp:</b></font> Achte bei der Umwandlung zu Markdown auf den linken Rand der Zelle: das <b>ln [ ] :</b> verschwindet.
</div>
<br>
 
### 2. Es gibt drei Arten, eine Markdown- in eine Code-Zelle zurückzuwandeln:
<br>
1. Du klickst im herunterklappenden Menü/Dropdown-Menü neben <b>Markdown</b> auf die grau hinterlegte, nach unten zeigende Pfeilspitze und wählst <b>Code</b> aus.
2. Du drückst in der neuen Zelle auf der Tastatur <font color = green>Esc</font> plus <font color = green>Y</font>.
3. Aus dem <b>Cell</b>-Auswahlmenü ganz oben wählst du <b>Cell Type</b> aus und dann <b>Code</b>.

 <div class="alert alert-block alert-info">
<font size="3"><b>Tipp:</b></font> Achte bei der Umwandlung zu Markdown auf den linken Rand der Zelle: das <b>ln [ ] :</b> erscheint.
</div>
<br>
  
### 3. Formatierungsmöglichkeiten im Markdown  
<br>
Du kannst in einer Markdown-Zelle nicht nur simplen Fließtext schreiben, sondern diesen auch mit Überschriften, Farben, Absätzen, Aufzählungen, Tabellen und vielem mehr formatieren. Ein paar Beispiele hierzu:  
<br>

* eine Überschrift beginnt mit <font color = green>#</font>, eine kleinere Überschrift mit <font color = green>##</font>, eine noch kleinere mit <font color = green>###</font>
* Aufzählungen beginnst du mit einem <font color = green>*</font> oder <font color = green>+</font> oder <font color = green>-</font>
* um etwas fett zu schreiben, schreibst du es innerhalb von zwei <font color = green>**</font> oder innerhalb von zwei <font color = green>__</font> oder du setzt es in folgenden HTML-Code: <font color = green>&lt;b></font>fett<font color = green>&lt;/b></font>   
<br>
<br>

### 4. Ein Ausflug zu den "Entkommungssequenzen"  
<br>
"Entkommungssequenzen", besser bekannt als Escape-Sequenzen, benötigst du, um Zeichen, die in der Sprache HTML vorkommen, lesbar im Markdown darzustellen. Da Markdown teils HTML-basierter Code ist, werden einige HTML-Codezeichen nicht als Fließtext dargestellt, zum Beispiel das "kleiner als"-Zeichen des HTML-Codes: <font color = green><</font>  
<br>
    
Im Markdown wird HTML-Code als Code gelesen - nicht als Text. Das bedeutet, nach Ausführung einer Zelle interpretiert Python die HTML-Zeichenfolgen und setzt sie um. Wie sie umgesetzt werden, hängt von ihrer dahinterliegenden Funktion ab.  
Um zum Beispiel das "kleiner als"-Zeichen und dessen nachfolgenden HTML-Code lesbar darzustellen, wird es ersetzt durch diese Escape-Sequenz: ``&lt;``  
<br>
<div class="alert alert-block alert-warning">
<font size="3"><b>Übung:</b></font>  
    
a) Verwandle die folgende Code-Zelle in eine Markdown-Zelle.  
    
b) Schreibe anschließend mit der HTML-Code-Variante ein Wort deiner Wahl fett. Führe die Zelle aus. Siehst du, dass der HTML-Code verschwunden ist?   

c) Schreibe darunter die fettende HTML-Code-Variante mit deinem Wort so, dass sie nach Ausführung der Zelle sichtbar ist. Hierzu brauchst du die oben genannte Ersetzung des "kleiner als"-Zeichens durch die Escape-Sequenz (wie in der Auflistung zu Wortfettungen oben, unter 3.).  

Beachte dabei, dass auf ein Wort, das nicht mehr innerhalb von HTML-Code, sondern mit einer Escape-Sequenz abgebildet ist, die Funktionalität des HTML-Codes nicht mehr greift. In diesem Beispiel hier wird das Wort nicht mehr gefettet. Denn mit Escape-Sequenzen wird die Ausführung des HTML-Codes unterbunden.
</div>    

HTML-Escape-Sequenzen beginnen üblicherweise mit einem Kaufmanns-Und (&, engl.: ampersand) und enden mit einem Semikolon. 

Der **Escape-Character** <font color = green>\\</font> (Backslash) funktioniert hingegen sprachübergreifend, so auch in Python. Möchtest du beispielsweise den Escape-Character <font color = green>\\</font> in einer Markdown-Zelle schreiben, setzt du einen zweiten Backslash davor: <font color = green>\\\\</font>, denn sonst wird der einzelne Backslash als Escape-Character interpretiert. 
<br>
Beispiele:
* <font color = green>\#</font> würde als Zeichen für eine Überschrift interpretiert werden, aber nicht mit: <font color = green>\\#</font>
* <font color = green>\*</font> würde als Zeichen für eine Aufzählung interpretiert werden, aber nicht mit: <font color = green>\\*</font>
* <font color = green>\+</font> würde als Zeichen für eine Aufzählung interpretiert werden, aber nicht mit: <font color = green>\\+</font>

<div class="alert alert-block alert-info">
    <font size="3"><b>Merke:</b></font> Mit Escape-Sequenzen (engl.: escape sequences)  entkommst du der Interpretation einer HTML-Zeichenfolge als Code. Mit ihnen stellst du Zeichenfolgen, die als HTML-Code interpretiert werden würden, lesbar dar.
</div>  
<br>

Es gibt viele Escape-Sequenzen. In einer späteren Einheit lernst du noch weitere kennen.
