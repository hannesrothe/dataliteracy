* ein Tuple ist unveränderlich => mmmh, wir können also keinen eingetragenen Wert mehr ändern. Was ist, wenn wir uns mal verschreiben und ein Wort falsch eingetragen ist? Wir könnten das nur rückgängig machen, indem wir das komplette Tuple neu übertragen. Das kann sehr umständlich werden!

    <font size="3"><b>Der Scope von Variablen:</b></font>  
    
Der <b>Scope</b> von Variablen ist ihr Geltungsbereich bzw. der Bereich, in welchem sie dem restlichen Programm bekannt sind.  
<br>

<br>

<br>

<br>

Eine Variable, die außerhalb von If-Statements, Loops und Funktionen deklariert wird, ist **global** bekannt. Das komplette Programm kennt sie.

<div class="alert alert-block alert-info">
<h3>Das kannst du aus dem Kapitel "3.5 Dictionaries" mitnehmen:</h3>

* **Dictionary**
    * ist ein weiterer Collection-Type von Python
    * ist wie die anderen Collection-Types ein Container, in dem Werte gesammelt werden
    * die Python-interne Bezeichnung dieses Datentyps lautet <b>dict</b>
    * besteht aus Key-Value-Pairs = Schlüssel-Werte-Paaren (im Gegensatz zu List)
        * Keys sind individuelle Indize, bestehend aus immutable (=unveränderlichen) Datentypen
        * Values können jeglichen Datentyps sein
    * hat keine feste Reihenfolge der Elemente (im Gegensatz zu List)
    * ist beliebig kürz- und erweiterbar
    * Definition eines leeren Dictionarys mit zwei geschweiften Klammern: ``{}``
    * Separierung von <b>Key</b> und <b>Value</b> mittels Doppelpunkt: <b>Key: Value</b> 
    * Trennung der Key-Value-Paare mit einem Komma (wie bei List), z.B.: ``dct = {'A': 1, 'B': 2, 'C': 3}``
    * Dictionarys benötigen mehr Speicherplatz als Lists, das Durchsuchen nach Werten verläuft aufgrund der Indexierung über die Keys jedoch schneller
    * **Hauptvorteil von Dictionarys: Werte können über die Keys genau bezeichnet werden**  
<br>
* **mutable (veränderliche) und immutable (unveränderliche) Datentypen**
    * jedes Objekt hat eine ID, einen Namen, einen Wert und einen Datentyp
    * bei veränderlichen Datentypen können diese Eigenschaften in-place verändert werden
    * bei unveränderlichen Datentypen ist eine in-place Änderung von mind. einer dieser Eigenschaften ausgeschlossen
    * von unveränderlichen Datentypen wird **keine flache** Kopie beim Kopieren und bei Änderungen erstellt
    * immutable Datentypen:
        * String
        * Integer
        * Float
        * Complex
        * Boolean
        * Tuple
        * Range
    * mutable Datentypen
        * List
        * Dictionary
        * Set
        * benutzerdefinierte Datentypen  
<br>
* **Erstellung eines Dictionarys**
    * über die oben beschriebene Syntax mit gescheiften Klammern, Doppelpunkt zur Trennung von <b>Key: Value</b> und Trennung der Einträge mit einem Komma, z.B.: ``dct = {'A': 1, 'B': 2, 'C': 3}``
        * Beispiel für Erstellung eines nested Dictionarys über die Dictionary-Syntax: 
            * ``nested_dict = {'s1': {'name': 'Reno Zak', 'age': 21}, 's2': {'name': 'Zakira Renoff', 'age': 25}}``
    * über die Funktion ``dict()``
     * Beispiel zur Erstellung eines neuen Dictionarys: 
         * ``student = dict(Name = 'Tedi Tedson', Semester=1)``
     * Beispiel zur Erstellung eines Dictionarys aus einer Liste mit Tupeln: 
         * ``studs_dct = dict(['Tedi Tedson', 1), ('Inga Igsun', 2)])``
     * Beispiel zur Erstellung eines nested Dictionarys über ``dict()``:
         * ``students = dict(s1 = {'name': 'Reno Zak', 'age': 21}, s2 = {'name': 'Zakira Renoff', 'age': 25})``
    * aus einer String-Liste über ``dict.fromkeys()``
         * Syntax: <font color = green>dict.fromkeys(stringlist, startwert)</font> 
         * die Strings der Liste werden zu Keys des Dictionarys mit dem angegebenen Startwert umgewandelt  
         * Beispiel zur Erstellung eines Dictionarys aus einer Liste von Strings:
             * ``lst = ['A', 'B', 'c']``
             * ``dct = dict.fromkeys(lst, 0)``
             * Output über ``print(dct)`` => {'A': 0, 'B': 0, 'C': 0}
    * aus zwei Listen, von denen eine die Keys und die andere die Values enthält über ``zip()``
        * Beispiel:
            * ``keys = ['A', 'B', 'C']``
            * ``vals = [1, 2, 3]``
            * ``combo_dct = dict(zip(keys, vals)`` => Output über ``print(combo_dct)``: {'A': 1, 'B': 2, 'C': 3}
        * ist eine Liste länger als die andere, werden die zu vielen Werte der längeren Liste weggelassen    
<br>
* **Wertezugriff und -überprüfung**
    * ähnlich zu Lists über eckige Klammern, jedoch statt einem Integer-Index über die Keys, z.B.: ``dct['k1']``
    * Slicing funktioniert **nicht** bei Dictionarys
    * Abruf aller Key-Value-Paare eines Dictionarys über die Funktion ``.items()``, z.B.: ``dct.items()``
    * Abruf aller Keys eines Dictionarys über die Funktion ``.keys()``, z.B.: ``dct.keys()``
    * Abruf aller Values eines Dictionarys über die Funktion ``.values()``, z.B.: ``dct.values()``
    * Abruf aller Values eines Subdictionarys über die Funktion ``.values()``, z.B.: ``dct['s1'].values()``
    * Abruf aller Keys eines Subdictionarys über die Funktion ``.keys()``, z.B.: ``dct['s1'].keys()``
    * Überprüfung, ob ein Key enthalten ist über den Membership-Operator ``in``, z.B.: ``'Semester' in student``
    * Überprüfung, ob ein Value enthalten ist über ``in`` und ``.values()``, z.B.: ``'1' in student.values()``    
    <br>
* **Autovervollständigung**
    * Variablennamen können autovervollständigt werden, was dir Schreibarbeit spart
    * Schritte:
        * schreibe mindestens den Anfangsbuchstaben der Variable
        * klicke die Tabulator-Taste
        * wähle den richtigen Namen mit den Pfeiltasten (oben/unten) oder dem Mausrad aus
        * bestätige mit Enter
    * je mehr Buchstaben der Variable du schreibst, desto weiter oben erscheint sie in der Auswahlliste
    * ist die Variable einzigartig unter den Auswahlmöglichkeiten, wird sie prompt über die Tab-Taste geschrieben (ohne extra Enter-Klick)  
<br>
* **Dictionary-Einträge updaten**
    * ``=`` (Zuweisungsoperator), um einen Wert an einem Key zu ändern oder ein Key-Value-Paar hinzuzufügen
        * Beisiel: ``dct['Schlüssel'] = 'Wert'``
        * der Wert an dem angegebenen Schlüssel wird überschrieben (Achtung!)
        * 'Schlüssel' und 'Wert' können auch ein ganz neues Key-Value-Paar sein, das damit dem Dictionary hinzugefügt wird
        * Der Key kann nicht geändert werden, es kann nur ein neuer hinzugefügt und der alte gelöscht werden
        * Zuweisung eines nested Dictionarys mit z.B.: ``dct['Schlüssel'] = {'K1': 'V1', 'K2': 'V2'}``
        * Wertänderung an einem nested Dictionary mit z.B.: ```dct['Schlüssel']['K1'] = 'B3'`` => 'V1' wird zu 'B3'      
    * ``.update()``, um Werte und Key-Value-Paare zu ändern und hinzuzufügen
        * Beispiel: ``dct.update(Schlüssel5 = 'Wert5')``
        * existiert **Schlüssel5** noch nicht, wird er mit seinem Wert hinzugefügt, ansonsten wird sein vorheriger Wert überschrieben (Achtung!)
        * Beispiel für die Zuweisung eines zuvor definierten Dictionarys: ''dct.update(dct2)
            * **dct2** wird dabei **kein** Subdictionary von **dct**
        * Beispiel für die Zuweisung eines Dictionarys **und** Key-Value-Paars: ``dct.update(dct2, K = 'V'``)
        * Beispiel für das Einfügen eines Subdictionarys: ``dct.update(Schlüssel = {'K1': 1, 'K2': 2})``
        * Update eines Subdictionarys mit z.B.: ``dct['Schlüssel'].update('K1': 55)`` => 'K1' wurde auf 55 geändert
            * zur Änderung eines Keys eines Subdictionarys muss das Subdictionary komplett, auch mit den restlichen Keys, neu übergeben werden, z.B.: ``dct['Schlüssel'].update({'S1': 1, 'K2': 2})`` => der Key 'K1' wurde zu 'S1' verändert
        * ist praktisch, um ein leeres Dictionary zu befüllen
        * verwende ``dict.fromkeys()`` niemals im Zusammenhang mit dieser Syntax von ``.update()``:</b> <font color = darkred>dct['Key1'].update({'Subkey1': 'Subvalue1'})</font>

* **Entfernen von Einträgen und von Dictionarys**
    * ``.pop()`` entfernt Key-Value-Paare und gibt den Wert des gelöschten Keys aus, z.B.: ``dct.pop('K1')`` => der Wert zu 'K1' erscheint in der Ausgabe
        * existiert der angegebene Key nicht, wirft Python einen <font color = darkred>KeyError</font>
    * ``.popitem()`` entfernt den zuletzt hinzugefügten Dictionary-Eintrag und gibt das gelöschte Key-Value-Paar aus, z.B.: ``dct.popitem()``

    * ``del`` zum Löschen von Key-Value-Paaren und Dictionarys
        * Beispiel für das Löschen eines Key-Value-Paars: ``del dct['K1']`` => löscht 'K1' und dessen Value
        * Beispiel für das Löschen eines Dictionarys: ``del dct``
        * über ``del`` erhält man **keine** Ausgabe darüber, was gelöscht wurde

* **Dictionarys vergleichen**
    * ``==`` zum Inhaltsvergleich
        * Vergleich der Key-Value-Paare zweier Dictionarys, z.B.: ``dct1 == dct2`` => Output ist True bei exakter Gleichheit
            * entspricht: ``dct1.items() == dct2.items()``
        * Vergleich der Keys zweier Dictionarys, z.B.: ``dct1.keys() == dct2.keys()``
        * Vergleich der Values zweier Dictionarys, z.B.: ``dct1.values() == dct2.values()``
        * Vergleich der Eintragsanzahlen zweier Dictionarys, z.B.: ``len(dict1) == len(dict2)``
    * ``-`` für die Unterschiede eines Dictionarys zu einem anderen (= Difference)
        * Beispiel für die Unterschiede von **dct1** zu **dct2**: ``dct1 - dct2`` => Output sind die Key-Value-Paare, die nur in dct1 vorkommen
        * Beispiel für die Unterschiede von **dct2** zu **dct1**: ``dct2 - dct1`` => Output sind die Key-Value-Paare, die nur in dct2 vorkommen
        * für den Vergleich der Keys ist ``.keys()``, für den Vergleich der Values ``.values()`` einzusetzen
    * ``&`` für die Gemeinsamkeiten zweier Dictionarys (= Intersection), z.B.: ``dct1.items() & dct2.items()`` => Output sind die Key-Value-Paare, welche sowohl in dct1, als auch in dct2 enthalten sind
        * **wichtig:** Für den Vergleich mit ``&`` muss ``.items()``, ``.keys()`` oder ``.values()`` angegeben werden, sonst kommt es zu einem <font color = darkred>TypeError</font>
    * **Vergleichsoperatoren** für den "Höhenvergleich" von Buchstaben **oder** Integers **oder** Floats
        * * **Wiederholung zu Vergleichsoperatoren**
            * liefern einen boolschen Rückgabewert (True oder False)
            * ``==``: prüft auf Gleichheit von Werten/Inhalten von Objekten
            * ``!=``: prüft Ungleichheit von Werten/Inhalten von Objekten
            * ``<``: prüft, ob der linke Wert kleiner als der rechte ist
            * ``>``: prüft, ob der linke Wert größer als der rechte ist
            * ``<=``: prüft, ob der linke Wert kleiner als der rechte ist oder gleich dem rechten
            * ``>=``: prüft, ob der linke Wert größer als der rechte ist oder gleich dem rechten  
        * Beispiel für den Vergleich von zwei Werten zweier Dictionarys: ``dct1['K1'] < dct2['S1']`` => prüft, ob der Wert von 'K1' kleiner als der Wert von 'S1' ist, wenn ja, Output: True
    * **Set-Funktionen** zum Vergleich von Dictionarys
        * hierfür müssen die Dictionarys zu Sets konvertiert werden, z.B.: ``st1 = set(dct1)``
            * diese Beispielkonvertierung konvertiert die Dictionary-Keys zu Set-Werten
            * Konvertierung der Key-Value-Paare zu Set-Einträgen in Form von Tupeln, z.B.: ``st1 = set(dct1.items())``
            * Konvertierung der Values zu Set-Einträgen, z.B.: ``st1 = set(dct1.values())``
        * ``.difference()``: Unterschiede des einen Sets zu dem anderen, z.B.: ``st1.difference(st2)`` => zeigt die Keys/Values/Key-Value-Paare (Tuples) an, die nur in st1 vorkommen
            * entspricht beim Vergleich der Key-Value-Paare: ``dct1 - dct2``
                * **Achtung: anzufügen an dct1 und dct2 sind ``.keys()``, ``.values()`` oder ``.items()``**
        * ``.symmetric_difference()``: alle Unterschiede zweier Sets, z.B.: ``st1.symmetric_difference(st2)`` => zeigt alle Werte, die sowohl nur im ersten und nur in dem zweiten Set vorkommen
            * entspricht beim Vergleich der Key-Value-Paare: ``(dct1 - dct2) + (dct2 - dct1)``
                * **Achtung:** anzufügen an dct1 und dct2 sind ``.keys()``, ``.values()`` oder ``.items()``
        * ``.intersection()``: was beide Sets gemeinsam haben, z.B.: ``st1.intersection(st2)`` => zeigt alle Werte, die sowohl im ersten, als auch im zweiten Set vorkommen
            * wenn die Sets aus den Dictionary-Keys bestehen, entspricht das diesem Abgleich der Dictionary-Keys: ``dct1.keys() & dct2.keys``
        * ``.union()``: welche Werte in beiden Sets - ohne Dopplungen - vorkommen, z.B.: ``st1.union(st2)`` 0> zeigt alle Werte von st1 und alle von st2, wobei Dopplungen überschrieben und damit eliminiert werden

* **mit sorted() Dictionarys sortieren**
    * Dictionarys sind grundsätzlich ungeordnet, Keys haben keine feste Position
    * die Sortierung von Dictionarys dient eher repräsentativen Zwecken, denn auch sortierte Dictionarys bleiben bei erneuten Aufrufen nicht sortiert
    * Wiederholung: 
        * ``sorted()`` liegt ein stabiler, **nicht** in-place Sortieralgorithmus zugrunde, der in aufsteigender Reihenfolge sortiert
        * Syntax: <font color = green>sorted(iterierbaresObjekt, key=None, reverse=False)</font>
    * der Rückgabewert - auch bei der Sortierung von Dictionarys - ist eine Liste
    * Beispiel für die Sortierung der Dictionary-Keys: ``sorted(dct1)`` => Output ist eine Liste mit geordneten Werten (den Keys des Dictionarys)
    * Beispiel für die Sortierung der Dictionary-Values: ``sorted(dct1.values())`` => Output ist eine Liste mit geordneten Werten (den Values des Dictionarys)
    * Beispiel für die Sortierung der Dictionary-Key-Value-Paare: ``lst1 = sorted(dct1.items())`` => Output über ``print(lst1)`` ist eine Liste mit nach den Keys geordneten Tupeln (die Key-Values des Dictionarys jeweils zu einem Tuple zusammengefasst)
        * die Rückumwandlung in ein sortiertes Dictionary ist möglich mit z.B.: ``dct1 = dict(lst1)``
    * Beispiel für die Sortierung eines Subdictionarys: 
        * ``subdct_sortedlst = sorted(dct1['Subdctkey'])`` => subdct_sortedlst enthält das Subdictionary als sortierte Liste
        * ``dct1['Subdctkey'] = dict(subdct_sortedlst)`` => dem Superdictionary wird die sortierte Liste, die zu einem Dictionary zurück konvertiert wurde, am Key des Subdictionarys zugewiesen (das zuvor unsortierte Subdictionary wird dabei überschrieben)
    * gleichzeitige Sortierung von Super- und Subdictionarys mit dem Modul **json** und der Funktion **json.dumps()**
        * **JSON** ist ein Transportformat zum Senden und Empfangen von Daten/Dateien, das im Web-Bereich stark verbreitet ist
        * **JSON** bedeutet: <b>J</b>ava<b>S</b>cript <b>O</b>bject <b>N</b>otation
        * Import über: ``import json``
        * Syntax: <font color = green>json.dumps(dict, sort_keys=True, indent=int)</font>  
        * Beispiel:
            * ``letters = dict(O = 15, ABC = dict(A = 1,  C = 3, B = 2))``
            * ``letters_complsorted = json.dumps(letters, sort_keys=True, indent=4)``
                * **letters** wird über ``json.dumps()`` und ``sort_keys=True`` nach seinen Keys sortiert, inkl. der Keys des Subdictionarys (nicht möglich mit Values)
                * optional: das Subdictionary wird über ``indent=4`` in der Ausgabe um 4 Leerzeichen eingerückt
        * mehr über das Python-Modul **json**, seine Umwandlungen und Funktionen, erfährst du hier: https://docs.python.org/2/library/json.html

* **Der Scope von Variablen**
    * ist ihr Bekanntheitsbereich im restlichen Programm
    * wird eine Variable **innerhalb** eines If-Statements, eines Loops oder einer Funktion deklariert, kennt sie das Programm außerhalb dieser Konstrukte **nicht** -> sie ist nur **lokal** in ihnen bekannt
    * eine Variable, die **außerhalb** eines If-Statements, eines Loops oder einer Funktion deklariert wird, ist dem kompletten Programm bekannt -> sie ist **global** bekannt und in allen anderen Programmteilen zugreif- und somit veränderbar
