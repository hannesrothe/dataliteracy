## 3.6.6 Nested For-Loop

In der Praxis wirst du nicht umhin kommen, auch mit verschachtelten Listen und Dictionarys zu arbeiten.  

Auch auf diese kannst du For-Loops und List-/Dictionary-Comprehensions anwenden, um über die tieferen Ebenen zu iterieren. Allerdings werden diese dann auch verschachtelt. Wie das funktioniert, siehst du jetzt an einem Beispiel.

### Beispiel zu einer verschachtelten Liste und einem verschachtelten For-Loop

Im Folgenden siehst du eine Liste mit Sublisten. Diese Datenstruktur wird auch Liste aus Listen bzw. List of Lists genannt. Eine weitere Bezeichnung für sie ist **Matrix**, denn sie ist wie eine Verzahnung verschiedener Schichten bzw. wie eine Tabelle.   

```python
list_of_lists = [['A', 1],
                 ['B', 2],
                 ['C', 3]]

# nested For-Loop:

for sublist in list_of_lists:
    for value in sublist:
        print(value)
```

#### Aufbau dieses verschachtelten For-Loops

``for sublist in list_of_lists:`` => Dieser Teil entspricht dem dir bekannten For-In-Statement eines For-Loops über eine Liste. Die Variable <b>sublist</b> hätte auch mit <b> i</b> oder anders benamt werden können. Zum besseren Verständnis wurde sie  <b>sublist</b> genannt, denn mit dem obersten For-In-Statement greifen wir auf die oberste Ebene der Liste zu: auf die Ebene, welche die Sublisten enthält. Die Variable <b>sublist</b> steht also für die aktuelle Subliste, über welche iteriert wird. **Man nennt diesen Teil auch "äußerer For-Loop".**  

``for value in sublist:`` => Das zweite For-In-Statement muss mit 4 Leerzeichen oder einem Tabulator-Abstand eingerückt werden. Nur so weiß Python, dass es dem oberen For-In-Statement untergeordnet ist.  
Um auf die Werte in den Sublisten zugreifen zu können, wird hier die Variable <b>sublist</b> als Iterable nach ``in`` eingesetzt. Für die einzelnen, aktuellen Werte in den Sublisten wird eine neue Variable, <b>value</b>, festgelegt. Auch diese könnte anders heißen. **Man nennt diesen Teil auch "innerer For-Loop".**      

``print(values)`` => Mit einer weiteren Einrückung von 4 Leerzeichen/1 Tabulator-Abstand (insgesamt 8 Leerzeichen bzw. 2 Tab-Abstände) sagst du Python, dass diese Funktion dem vorhergehenden For-In-Statement untergeordnet ist. Das muss sie auch sein, denn der <b>value</b>, den es auszugeben gilt, wird erst in dem zweiten For-In-Statement definiert.  
<br>

#### Ablauf der Iteration im verschachtelten For-Loop

Der nested For-Loop sieht für dich vielleicht auf den ersten Blick komplizierter aus, als er ist. Er ist eine Erweiterung zum dir bekannten For-Loop und funktioniert auf den unteren Ebenen so, wie auf der obersten.  
Gehe am besten den folgenden Ablauf gedanklich durch, damit dir das Muster im Ablauf des nested For-Loops klar wird.  

Hier ist noch einmal das Beispiel von oben:  

``list_of_lists = [['A', 1],
                 ['B', 2],
                 ['C', 3]]``

``for sublists in list_of_lists:
    for value in sublists:
        print(value)``

**1. Iteration:**  
* 1. For-In-Statement: Das 1. Element der <b>list_of_lists</b> wird angewählt: ``['A', 1]`` auf Index [0].
* 2. For-In-Statement: Das 1. Element der 1. Subliste wird angewählt: ``'A'`` auf Index [0][0].
* Über ``print(value)`` wird der im For-Loop aktuelle Wert ausgegeben.  


**2. Iteration:**  
* 1. For-In-Statement: wird gerade nicht weiter geführt, weil der For-Loop noch nicht mit der Iteration des 2. For-In-Statements fertig ist.
* 2. For-In-Statement: Das 2. Element der 1. Subliste wird angewählt: ``1`` auf Index [0][1].
* Über ``print(value)`` wird der im For-Loop aktuelle Wert ausgegeben.  
=> die Iteration über die erste Subliste ist abgeschlossen. Deshalb geht der For-Loop automatisch mit einer Weiterzählung im 1. For-In-Statement weiter.


**3. Iteration:**  
* 1. For-In-Statement: Das 2. Element der <b>list_of_lists</b> wird angewählt: ``['B', 2]`` auf Index [1].
* 2. For-In-Statement: Das 1. Element der 2. Subliste wird angewählt: ``'B'`` auf Index [1][0].
* Über ``print(value)`` wird der im For-Loop aktuelle Wert ausgegeben.  

**4. Iteration:**  
* 1. For-In-Statement: wird gerade nicht weiter geführt, weil der For-Loop noch nicht mit der Iteration des 2. For-In-Statements fertig ist.
* 2. For-In-Statement: Das 2. Element der 2. Subliste wird angewählt: ``2`` auf Index [1][1].
* Über ``print(value)`` wird der im For-Loop aktuelle Wert ausgegeben.  
=> die Iteration über die zweite Subliste ist abgeschlossen. Deshalb geht der For-Loop automatisch mit einer Weiterzählung im 1. For-In-Statement weiter.

**5. Iteration:**  
* 1. For-In-Statement: Das 3. Element der <b>list_of_lists</b> wird angewählt: ``['C', 3]`` auf Index [2].
* 2. For-In-Statement: Das 1. Element der 3. Subliste wird angewählt: ``'C'`` auf Index [2][0].
* Über ``print(value)`` wird der im For-Loop aktuelle Wert ausgegeben.  

**6. Iteration:**  
* 1. For-In-Statement: wird gerade nicht weiter geführt, weil der For-Loop noch nicht mit der Iteration des 2. For-In-Statements fertig ist.
* 2. For-In-Statement: Das 2. Element der 2. Subliste wird angewählt: ``3`` auf Index [2][1].
* Über ``print(value)`` wird der im For-Loop aktuelle Wert ausgegeben.  
=> die Iteration über die dritte Subliste ist abgeschlossen. Deshalb geht der For-Loop automatisch mit einer Weiterzählung im 1. For-In-Statement weiter. Doch weil auch hier alle Indize der <b>list_of_lists</b> durchiteriert wurden, stoppt der nested For-Loop an dieser Stelle endgültig.  
<br>

<div class="alert alert-block alert-info">
<font size="3"><b>Tipp zum Grundschema von nested For-Loops:</b></font> 
<br>

Dieses Schema ist für alle nested For-Loops über alle Arten von Iterables gleich:
* Das letzte For-In-Statement wird solange ausgeführt, bis die Iteration über dessen Iterable-Elemente abgeschlossen ist. 
* Danach zählt das darüber liegende For-In-Statement eine Position weiter. 
* Das passiert solange, bis die Iteration des obersten und anschließend innersten For-In-Statements abgeschlossen ist.

Das gilt auch für mehr als zwei Verschachtelungen von For-Loops.
</div>
<br>
<br>

An dem folgenden Beispiel kannst du das nachvollziehen. Zuerst ist der äßere Loop aktiv, danach wird der innere Loop vollständig abgearbeitet. Danach zählt der äußere Loop eine Iteration weiter und wieder wird der innere Loop vollständig abgearbeitet. Das geht solange, bis die letzte äußere und anschließend die letzte innere Iteration abgelaufen sind:  

```python
for i in range(1,4):
    print(f'Äußerer Loop ist zum {i}. Mal aktiv')
    for k in range(1,3):
        print(f'Innerer Loop ist zum {k}. Mal aktiv')
```

### Eine Matrix mit range() und einem nested For-Loop erstellen

In der Datenverarbeitung wird es mitunter nötig sein, Matrizen mit Zahlenwerten anzulegen. Der Vorteil hierbei ist, dass diese als Tabellen verwendet werden können, in welchen jeder Wert auf festen Indizen abgespeichert ist.  
Im Bereich "Machine Learning" werden sie für Berechnungen verwendet, die das Eintreffen bestimmter Ereignisse vorhersagen oder welche die Klassifizierung verschiedener Ergebnisse ermöglichen.  
<br>

#### Beispiel zu einer nested List, die mit range() und einem nested For-Loop erstellt wird

Erinnerung an die Syntax von ``range()``: <font color = darkgreen>range(start,stop,step)</font>

```python
nested_list = []

for y in range(3):
    nested_list.append([])
    for x in range(2):
        nested_list[y].append(x)

nested_list
```

**Erklärung:**  

``nested_list = []`` => Zunächst einmal muss eine leere Liste definiert werden. Diese wird anschließend im nested For-Loop befüllt.  

``for y in range(3):`` => Durch den 1. For-Loop-Body wird mit ``range(3)`` drei Mal iteriert. **Du kannst dir diesen Teil als Y-Achse in einem Koordinatensystem vorstellen. Hier wird die Höhe/Zeilenanzahl der Hauptliste festgelegt.** 

``nested_list.append([])`` => Drei Mal wird die angelegte Liste auf ihren Indizen von 0-2 (wegen ``range(3)``) mit jeweils einer leeren Liste befüllt. Das sind die Sublisten auf der 2. Ebene der angelegten Liste. Jede der drei Zeilen auf der Y-Achse enthält damit eine leere Subliste.  

``for x in range(2):`` => Hier befinden wir uns auf der 2. Ebene der angelegten Liste: auf einer der zuvor angelegten leeren Sublisten. Durch diese wird mit ``range(2)`` zwei Mal iteriert, von 0 bis 1. <b>x</b> nimmt durch ``range(2)`` die Werte 0 und 1 an. **Diesen Teil kannst du dir als X-Achse in einem Koordinatensystem vorstellen. Hier wird die Breite/Spaltenanzahl der Hauptliste festgelegt.**  

``nested_list[y].append(x)`` => Hier verwenden wir die mit ``range()`` erzeugten Integers über die Variable <b>x</b> (0 und 1). Über das <b>y</b> wird mit ``nested_list[y]`` die in der Iteration aktuelle Subliste angewählt. In diese wird mit ``.append(x)`` die aktuelle Integer aus ``range(2)`` eingefügt. ``range(2)`` zählt dabei von 0-1, weshalb die 0 am 1. Index der Subliste steht, die 1 am 2. Index.  
<br>
<div class="alert alert-block alert-info">
<font size="3"><b>Tipp zum Aufbau von Listen über nested For-Loops:</b></font> 
<br>

Das äußere/erste For-In-Statement steht für die Höhe/Zeilen (engl. rows) der Tabelle bzw. für die Y-Achse in einem Koordinatensystem.  

Das innere/zweite For-In-Statement steht für die Breite/Spalten (engl. columns) der Tabelle bzw. für die X-Achse in einem Koordinatensystem.
</div>
<br>
<br>

**Das Folgende funktioniert nur mit 2d Listen: Noch deutlicher wird dir das Prinzip mit der Y- und X-Achse in Relation zum äußeren und inneren For-In-Statement, wenn du dir mit der Python-Library numpy die Liste als Matrix/Tabelle ausgeben lässt:**

```python
import numpy as np

nested_list = []

for y in range(3):
    nested_list.append([])
    for x in range(2):
        nested_list[y].append(x)

# Anwendung der Numpy-Funktion .matrix()
print(np.matrix(nested_list))
```

Das Modul <b>numpy</b> wird wie gewohnt mit ``import`` importiert. Um nicht bei jedem Funktionszugriff auf das Modul den vollen Namen <b>numpy</b> ausschreiben zu müssen, wird mit ``as np`` Python gesagt, dass <b>np</b> jetzt der Name für numpy ist.  
Über ``np.`` wird auf die Numpy-Funktionen zugegriffen. Du könntest sie an dieser Stelle über die Tab-Taste durchblättern. Die Funktion ``.matrix()`` stellt Listen als Tabellen dar. Mit ihr siehst du, wie eine Liste als Tabelle tatsächlich aussehen würde - nur ohne die gewohnten Abgrenzungen über Linien.  

An der Ausgabe der oberen Code-Zelle siehst du, dass die <b>nested_list</b> aus drei Zeilen (Y-Achse) und zwei Spalten (X-Achse) besteht.

### Nested For-Loops mit Dictionarys

Nested For-Loops mit Dictionarys funktionieren analog zu nested For-Loops über Listen sowie andere Iterables, nur dass wir für Dictionarys - für deren Key-Value-Paare - wieder zwei Variablen benötigen. Nicht zu vergessen ist auch die Dictionary-Funktion ``.items()``, mit deren Hilfe wir die Keys und Values abrufen.   

#### Beispiel zu einem nested For-Loop über ein Dictionary

Das folgende Dictionary hat 2 Verschachtelungsebenen. Es enthält die Namen deiner KundInnen, wie viele Punkte sie gesammelt und wie viele NeukundInnen sie für dich angeworben haben.

```python
cust_accounts = {'Margherita Seppi': {'Punkte': 35, 'Akquiriert': 1},
                 'Lizzy Hagen': {'Punkte': 53, 'Akquiriert': 2},
                 'Kito Zaleski': {'Punkte': 79, 'Akquiriert': 0},
                 'Mehrdad Amid': {'Punkte': 44, 'Akquiriert': 1}}

for k1,v1 in cust_accounts.items():
    print(f'{k1}:')
    for k2,v2 in v1.items():
        print(f'{k2}: {v2}')
```

**Erklärung:**

``for k1,v1 in cust_accounts.items():`` => Im 1. For-In-Statement wird über die 1. Ebene des Dictionarys iteriert. Auf dieser Ebene sind die Namen (<b>k1</b>) der KundInnen gespeichert sowie das Subdictionary (<b>v1</b>) einer/s jeden KundIn.  

``print(f'{k1}: ')`` => Dieser Teil muss wieder um 4 Leerzeichen/1 Tab-Abstand eingerückt werden. Hier werden die KundInnennamen mit einem Doppelpunkt ausgegeben.  

``for k2,v2 in v1.items():`` => Das findet auf der gleichen Einrückungsstufe wie die vorhergehende Print-Funktion statt, denn beide sind gleichwertig dem 1. For-In-Statement untergeordnet. Für den Zugriff auf die Subdictionarys werden zwei **neue** Variablen (<b>k2</b> und <b>v2</b>) für deren Keys und Values benötigt. Das Iterable nach ``in`` ist hier das Subdictionary aus dem 1. For-In-Statement: <b>v1</b>. Über dieses wird hier iteriert.  

``print(f'{k2}: {v2} ')`` => Hier werden nun die Keys (mit Doppelpunkt) und Values der Subdictionarys ausgegeben. Beachte die weitere Einrückungsstufe um 4 Leerzeichen/1 Tab-Abstand. Ohne diese würde es zu einem Fehler kommen, denn <b>k2</b> und <b>v2</b> wurden erst im übergeordneten, 2. For-In-Statement definiert.  
<br>

<div class="alert alert-block alert-info">
<font size="3"><b>Extra Tipp zur Formatierung:</b></font> 
<br>

Falls dir die Art der Ausgabe nicht gefällt, in welcher alle Dictionary-Elemente auf je eine neue Zeile gesetzt wurden, gibt es einen Trick:  

**Mit dem Parameter ``end = ''`` kannst du Elemente auf der gleichen Zeile erscheinen lassen. Er wird nach dem String (oder einer Variable) - mit einem Komma getrennt - in die Print-Funktion eingesetzt.**  

Er funktioniert so:  

* ``end = ''`` (mit leerem String) am Ende in der Print-Funktion: setzt das nächste Element auf die gleiche Zeile
* ``end = ' '`` (mit einem Leerzeichen als String) am Ende der Print-Funktion: gibt die Elemente auf der gleichen Zeile aus und lässt zusätzlich einen Abstand von einem Leerzeichen zwischen den Elementen (im unteren Beispiel zwischen der Punkte-Anzahl und "Akquiriert")
</div>
<br>

```python
# zusätzlich wurde in die 1. Print-Funktion vor den Namen "\n" zur Separierung der KundInnen auf neue Zeilen eingefügt

cust_accounts = {'Margherita Seppi': {'Punkte': 35, 'Akquiriert': 1},
                 'Lizzy Hagen': {'Punkte': 53, 'Akquiriert': 2},
                 'Kito Zaleski': {'Punkte': 79, 'Akquiriert': 0},
                 'Mehrdad Amid': {'Punkte': 44, 'Akquiriert': 1}}

for k1,v1 in cust_accounts.items():
    print(f'\n{k1}: ', end = '')
    for k2,v2 in v1.items():
        print(f'{k2}: {v2}', end = ' ')
```

### Nested For-Loops mit Bedingungen

Ebenso wie in "einfache" For-Loops, können Bedingungen auch in verschachtelte For-Loops gesetzt werden.
Sehen wir uns das anhand einer mit Integers gefüllten Liste an.  

#### Beispiel zu einem nested For-Loop mit einer Bedingung

Die Liste zeigt das Ranking von ComputerspielerInnen in Einzel- und Teamspielen. 1 ist die schlechteste Wertung, 10 die beste. Haben sie eine Wertung, die 7 oder höher ist, erscheint die Meldung: "SpielerIn [Listenposition] ist ein/e ziemlich gute/r SpielerIn."

```python
ranking = [[4, 7], [8, 9], [6, 5], [7, 9]]

for sublist in ranking:
    for i in sublist:
        if i >= 7:
            print(f'SpielerIn {ranking.index(sublist)} ist ein/e ziemlich gute/r SpielerIn.')
            break
```

**Erklärung:**

``for sublist in ranking:
    for i in sublist:`` => Die ersten beiden For-In-Statements sind analog zum ersten Beispiel in dieser Einheit. Mit dem 1. Statement wird über die oberste Ebene der Liste iteriert, welche die Sublisten enthält. Mit dem 2. Statement wird über die darunterliegende Listenebene iteriert. Hier wird auf die Werte innerhalb der einzelnen Sublisten zugegriffen.  
    
``if i >= 7:`` => Die Werte in den Sublisten werden darauf geprüft, ob sie größer oder gleich 7 sind.  

``print(f'SpielerIn {ranking.index(sublist)} ist ein/e ziemlich gute/r SpielerIn.')`` => Beachte die Einrückung dieser Bedingung um weitere 4 Leerzeichen/1 Tab-Abstand. Denn diese Bedingung ist dem inneren For-Statement untergeordnet (wegen der Definition der Variable <b>i</b> in ihm).  
Ist ein Sublisten-Wert größer/gleich 7, erfolgt diese Ausgabe.

``{ranking.index(sublist)}`` =>  Als Argument für die Ausgabe wurde dies zusätzlich  eingefügt. Hier wird mittels der im 1. For-In-Statement festgelegten Variablen und der Funktion ``.index()`` der aktuelle Index in der obersten Listenebene angezeigt. Denn dort, wo der Index der Subliste liegt, ist die Position der/s aktuellen Spielerin/s in der Hauptliste <b>ranking</b>.  

``break`` => Dieses Statement ist ebenfalls dem 2. For-In-Statement untergeordnet. Es ist insbesondere gleichrangig zum Print-Befehl. Deshalb muss es auch um insgesamt 8 Leerzeichen/2 Tab-Abstände eingerückt werden.  
Ohne dieses Statement würde die Print-Ausgabe zwei Mal erfolgen, wenn der Spieler in beiden Kategorien (Einzel- und Teamspiel) Punkte größer/gleich 7 hat. Auf diese Weise wird der innere For-Loop unterbrochen, sobald die Ausgabe über ``print()`` erfolgt ist. Danach geht die Iteration beim 1. For-In-Statement weiter, sprich: zur nächsten Subliste/zum nächsten Spieler.  

**In der folgenden Übung kannst du austesten, ob du verstanden hast, welche For-In-Statement-Variable auf was in der Liste zugreift.**

### Mehr als zweifach verschachtelte For-Loops

Je mehr Verschachtelungsstufen ein Iterable hat, desto mehr brauchst du in einem For-Loop, um auf sie zuzugreifen.  

>Zweifach verschachtelte Listen nennt man auch **2d Listen**, dreifach verschachtelte auch **3d Listen** usw. Im Internet wirst du statt List auch oft Array lesen. Das ist in Python ein und dieselbe Datenstruktur. In anderen Programmiersprachen wie Java haben Listen und Arrays unterschiedliche Eigenschaften und sind getrennte Datenstrukturen.

#### Beispiel zu einer dreifach erschachtelten Liste und einem dreifachen For-Loop

Wir können nicht nur zweifach verschachtelte Listen erstellen, sondern auch drei- und noch mehr verschachtelte. 

Das folgende Beispiel zeigt, wie mit einem dreifachen For-Loop eine 3d Liste erstellt wird. Das funktioniert wie die Erstellung eines 3d Koordinatensystem, welches eine Y-, X- und Z-Achse hat.   

```python
y = 2
x = 3
z = 4

lst = []

for i in range(y):
    lst.append([])
    for j in range(x):
        lst[i].append([])
        for k in range(z):
            lst[i][j].append(0)
            
lst
```

**Erklärung:**

``y = 2
x = 3
z = 4`` => Die Variablen für die Achsen bzw. die Tabellengrößen werden festgelegt.

``lst = []`` =>  Hier wird eine leere Liste definiert. Das ist die Hauptliste, die im nested For-Loop befüllt wird.  

``for i in range(y):`` => <b>i</b> ist die Variable für den aktuellen Wert an den Indizen der Hauptliste. Mit ``range(y)`` werden die ersten 2 dieser Indize angewählt - das ist die Höhe/Zeilenanzahl der Liste. Gleichzeitig enthält die Variable <b>i</b> die Werte 0 und 1, weil diese über ``range(y)`` = ``range(2)`` erzeugt wurden.   

``lst.append([])`` => Auf jeden dieser 2 Indize wird eine leere Liste gesetzt bzw. ein Platzhalter für die noch zu befüllenden weiteren Achsen.  

``for j in range(x):`` => Mit <b>j</b> wird zuerst die Variable für die aktuelle Position auf der X-Achse festgelegt. Sie erhält mit ``range(x)`` eine Länge von 3.

``lst[i].append([])`` => Hier wird drei Mal eine leere Liste pro Spalte (X-Achsen-Position) erstellt. Das ergibt 2x3 leere Listen.    

``for k in range(z):`` => <b>k</b> ist die Variable, welche über die aktuelle Zeilen- und Spaltenposition iteriert. Du kannst dir das wie bei dem Spiel "Schiffeversenken" vorstellen: Du nennst eine Position auf der Höhe (Y) und Breite (X), um dort hoffentlich ein Schiff zu treffen. Nur hier wird an dieser Position mit ``range(z)`` = ``range(4)`` vier Mal iteriert. Und weil in den vorhergehenden For-In-Statements 2x3 Mal iteriert wird, wird im gesamten nested For-Loop 2x3x4 Mal iteriert.

#### Zum besseren Verständnis siehst du hier ein Codebeispiel zur Arbeitsweise eines dreifach verschachtelten For-Loops

```python
for i in range(1,3):
    print(f'Äußerer Loop ist zum {i}. Mal aktiv')
    for j in range(1,3):
        print(f'Erster innerer Loop ist zum {j}. Mal aktiv')
        for k in range(1,3):
            print(f'Innerster Loop ist zum {k}. Mal aktiv')
```

<div class="alert alert-block alert-info">
<font size="3"><b>Tipp zum generellen Ablauf eines dreifach verschachtelten For-Loops:</b></font> 
<br>
    
**Generell wird nach der ersten Aktivierung des äußeren und danach nächst-inneren Loops:**   
1) Der innerste Loop vollständig abgearbeitet.  
2) Danach zählt der darüber liegende Loop 1 weiter und der innerste Loop wird erneut vollständig abgearbeitet.   
3) Das geschieht solange, bis der darüber liegende, erste innere und anschließend innerste Loop vollständig abgearbeitet sind.  
4) Danach zählt der äußere Loop eine Iteration weiter und das Schema beginnt wieder bei 1), bis alle Loops vollständig abgearbeitet sind.  

**Dieses Schema ist erweiterbar auf For-Loops mit mehr als drei Verschachtelungen.**

* **Nested For-Loops über Iterables ohne Key-Value-Paare**
    * Beispiel:
        * ``list_of_lists = [['A', 1], ['B', 2], ['C', 3]]``
        * ``for sublist in list_of_lists:``
            * ``for value in sublist:``
                * ``print(value)``
        * Output (jeweils untereinander): A 1 B 2 C 3
    * der äußerste (oberste) For-Loop iteriert stets über die äußerste Ebene des Iterables => in diesem Beispiel über die Hauptindize der <b>list_of_lists</b>, auf denen die Sublisten liegen
    * der äußerste For-Loop steht für die Y-Achse (Höhe) in einem Koordinatensystem bzw. für die Anzahl der Zeilen einer Tabelle
    * der darunterliegende, innere For-Loop steht für die X-Achse (Breite) in einem Koordinatensystem bzw. für die Anzahl der Spalten einer Tabelle
    * der innere Loop und alles zu ihm gehörige muss um weitere 4 Leerzeichen/1 Tab-Abstand weiter eingerückt werden
    * vom äußeren For-Loop wird vom ersten Hauptindiz (<b>sublist</b>) aus auf die Indize der Elemente (<b>value</b>) vom inneren For-Loop aus zugegriffen
    * tabellarische Ansicht einer **2d** Liste (zweifach verschachtelte Liste) ist möglich über:
        * ``import numpy as np``
        * ``print(np.matrix(nested_2dlist))``
    * **grundsätzlicher Ablauf (auch mehr als zweifach) verschachtelter For-Loops:**
        * das letzte For-In-Statement wird solange ausgeführt, bis die Iteration über dessen Iterable-Elemente abgeschlossen ist
        * danach zählt das darüber liegende For-In-Statement eine Position weiter
        * das passiert solange, bis die Iteration des obersten und anschließend innersten For-In-Statements abgeschlossen ist   
<br>
* **Erstellung einer Matrix mit range() und einem nested For-Loop**
    * Syntax von ``range()``: <font color = darkgreen>range(start,stop,step)</font>
    * Tipp zur Anzeige der mit ``range()`` erzeugten Werte: ``print(list(range()))``
    * ohne Angabe weiterer Argumente wird bei z.B. ``range(2)`` ein Range-Objekt von 0 bis 1 erstellt
    * Beispiel:
        * ``list_of_lists = [['A', 1], ['B', 2], ['C', 3]]``
        * ``nested_list = []``
        * ``for y in range(3):``
            * ``nested_list.append([])``
            * ``for x in range(2):``
                * ``nested_list[y].append(x)``
        * ``nested_list``
        * Output: [[0, 1], [0, 1], [0, 1]]
        * Erklärung: Im äußeren For-Loop werden drei leere Sublisten angelegt (entlang der Y-Achse/Höhe). Danach werden diese jeweils mit den Werten 0 und 1 befüllt (entlang der X-Achse/Breite)         
<br>
* **Nested For-Loops über Dictionarys**
    * Beispiel:
        * ``cust_accounts = {'Lizzy Hagen': {'Punkte': 53, 'Akquiriert': 2}, 'Kito Zaleski': {'Punkte': 79, 'Akquiriert': 0}}``
        * ``for k1,v1 in cust_accounts.items():``
            * ``print(f'\n{k1}: ', end = '')``
            * ``for k2,v2 in v1.items():``
                * ``print(f'{k2}: {v2}', end = ' ')``
            * Output (ohne Kreise zu Beginn):
                * Lizzy Hagen: Punkte: 53 Akquiriert: 2
                * Kito Zaleski: Punkte: 79 Akquiriert: 0

Dort wird der Key auf jeweils einer neuen Zeile ausgegeben und mit <b>, end = ''</b> wird das nachfolgende Element (der Key der Subliste: 'Punkte' mit seinen jeweiligen Values) auf der gleichen Zeile ausgegeben. Im inneren For-Loop wird auf die Keys und Values (<b>k2,v2</b>) der Subdictionarys  (<b>v1</b>) zugegriffen. Ausgegeben werden diese sowie ihre Values über <b>, end = ' '</b> mit einem Leerzeichen-Abstand zwischen 'Punkte' und 'Akquiriert' auf der gleichen Zeile.  
<br>
* **Nested For-Loops mit Bedingungen**
    * Beispiel:
        * ``for sublist in matrix:``
            * ``for i in sublist:``
                * ``if i >= 7:``
                    * ``print(i)``
        * Output: Alle Werte in den Sublisten, die größer/gleich 7 sind, werden ausgegeben
    * **Generell zu Bedingungen in Loops:**
        * diese müssen um weitere 4 Leerzeichen/1 Tab-Abstand unterhalb des zu ihnen gehörigen For-In-Statements eingerückt werden
        * je weiter die Hierarchie der Conditions reicht, desto weiter werden sie eingerückt
